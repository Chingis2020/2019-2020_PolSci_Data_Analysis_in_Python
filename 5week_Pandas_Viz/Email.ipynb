```python
import pandas as pd
import re
```

```python
data = pd.read_excel('список_История коррупции в России_Поросенков.xls', dtype= str)
data
```

```python
for i in data['11. Личные контакты']:
    #mails = re.findall(r'\b[\w\d\S]+@\w+.\w+', str(i))
    mails = re.findall(r'([a-zA-Z0-9_.+-]+)@([a-zA-Z0-9-]{2,})\.([a-zA-Z0-9-.]{2,})', str(i))
    print(mails)
```
